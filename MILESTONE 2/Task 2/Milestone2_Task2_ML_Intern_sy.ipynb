```python
# 1. Line Graph: Average Budget vs. Year

avg_budget_per_year = df.groupby('year')['budget'].mean()
plt.figure(figsize=(14, 6))
plt.plot(avg_budget_per_year.index, avg_budget_per_year.values, marker='o')
plt.title('Average Budget vs. Year')
plt.xlabel('Year')
plt.ylabel('Average Budget')
plt.grid(True)
plt.show()
```

```python
# 2. Bar Chart: Average Revenue vs. Year

avg_revenue_per_year = df.groupby('year')['revenue'].mean()
plt.figure(figsize=(18, 8))
avg_revenue_per_year.plot(kind='bar')
plt.title('Average Revenue vs. Year')
plt.xlabel('Year')
plt.ylabel('Average Revenue')
plt.grid(True)
plt.show()
```

```python
# 3. Scatter Plot: Budget vs. Revenue

plt.figure(figsize=(14, 6))
sns.scatterplot(data=df, x='budget', y='revenue')
plt.title('Budget vs. Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

```python
# 4. Pie Chart: Genre Distribution

genre_count = df['genres'].str.split(',').explode().value_counts()
plt.figure(figsize=(20, 14))
genre_count.plot(kind='pie', autopct='%1.1f%%')
plt.title('Genre Distribution')
plt.ylabel('')
plt.show()
plt.show()
```

```python
# 4. Bar Chart for Genre Distribution

# Calculate genre counts
genre_count = df['genres'].str.split(',').explode().value_counts()

# Create a bar chart
plt.figure(figsize=(12, 12))
sns.barplot(x=genre_count.values, y=genre_count.index, palette='viridis')
plt.title('Genre Distribution')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

```python
# 5. Histogram: Distribution of Movie Runtime
plt.figure(figsize=(14, 6))
sns.histplot(df['runtime'].dropna(), bins=20, kde=True)
plt.title('Distribution of Movie Runtime')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 6. Count Plot: Number of Movies per Year
plt.figure(figsize=(18, 6))
sns.countplot(x='year', data=df)
plt.title('Number of Movies per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# 7 Bar Chart for Average Revenue by Genre

# Assuming genres are separated by commas, we need to split and explode them
df_genre_revenue = df[['genres', 'revenue']].dropna()
df_genre_revenue['genres'] = df_genre_revenue['genres'].str.split(',')
df_genre_revenue = df_genre_revenue.explode('genres').reset_index(drop=True)

# Group by genres and calculate mean revenue
genre_revenue = df_genre_revenue.groupby('genres')['revenue'].mean().sort_values()

# Plot
plt.figure(figsize=(12, 12))
genre_revenue.plot(kind='barh', color='brown')
plt.title('Average Revenue by Genre')
plt.xlabel('Average Revenue')
plt.ylabel('Genre')
plt.grid(True)
plt.show()
```

```python
# 8 Bar Chart for Production Country vs Count of Movies

# Assuming the DataFrame df is already loaded and processed
# Split the production_countries column and explode the list
df_countries = df[['production_countries']].dropna()
df_countries['production_countries'] = df_countries['production_countries'].str.split(',')
df_countries = df_countries.explode('production_countries').reset_index(drop=True)

# Strip any leading/trailing whitespace from country names
df_countries['production_countries'] = df_countries['production_countries'].str.strip()

# Group by production country and count the number of movies
country_count = df_countries['production_countries'].value_counts()

# Plot bar chart
plt.figure(figsize=(18, 6))
sns.barplot(x=country_count.index, y=country_count.values, palette='viridis')
plt.title('Count of Movies in Each Production Country')
plt.xlabel('Production Country')
plt.ylabel('Count of Movies')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python

# 9 Bar Chart for Average Budget by Genre

# Assuming genres are separated by commas, we need to split and explode them
df_genre_budget = df[['genres', 'budget']].dropna()
df_genre_budget['genres'] = df_genre_budget['genres'].str.split(',')
df_genre_budget = df_genre_budget.explode('genres').reset_index(drop=True)

# Group by genres and calculate mean revenue
genre_budget = df_genre_budget.groupby('genres')['budget'].mean().sort_values()

# Plot
plt.figure(figsize=(12, 12))
genre_revenue.plot(kind='barh', color='green')
plt.title('Average budget by Genre')
plt.xlabel('Average budget')
plt.ylabel('Genre')
plt.grid(True)
plt.show()
```

```python
# 10 Violin plot : for Popularity vs. Genre

# Violin plot is useful for comparing the distribution of the data across different categories, and it will show the density of the popularity scores for each genre.

# Split genres and explode into a long format DataFrame
df_exploded = df.assign(genres=df['genres'].str.split(',')).explode('genres')

# Create a violin plot
plt.figure(figsize=(14, 8))
sns.violinplot(x='popularity', y='genres', data=df_exploded, palette='viridis', scale='width')
plt.title('Popularity Distribution by Genre')
plt.xlabel('Popularity')
plt.ylabel('Genre')
plt.show()
```

```python
# 11
# Scatter Plot: Popularity vs. Budget

plt.figure(figsize=(14, 6))
sns.scatterplot(x='budget', y='popularity', data=df)
plt.title('Popularity vs. Budget')
plt.xlabel('Budget')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

```python
# 12
# Scatter Plot: Popularity vs. Revenue

plt.figure(figsize=(14, 6))
sns.scatterplot(x='revenue', y='popularity', data=df)
plt.title('Popularity vs. Revenue')
plt.xlabel('Revenue')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

```python
# 13 Create a box plot for runtime
plt.figure(figsize=(10, 7))
plt.boxplot(df['runtime'].dropna(), vert=False, patch_artist=True)
plt.title('Distribution of Movie Runtimes')
plt.xlabel('Runtime (minutes)')
plt.show()
```

```python
# 14 Horizontal Bar Charts with Year Labels on Top
# Find the highest revenue movie for each year

highest_revenue_each_year = df.loc[df.groupby('year')['revenue'].idxmax()][['year', 'movie_title', 'revenue']]

# Plotting the highest revenue movie each year
plt.figure(figsize=(12, 20))
bars = plt.barh(highest_revenue_each_year['movie_title'], highest_revenue_each_year['revenue'], color='darkgreen')
plt.title('Highest Revenue Movie Each Year')
plt.xlabel('Revenue')
plt.ylabel('Movie Title')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest revenue movie at the top

# Add year labels on top of the bars
for bar, year in zip(bars, highest_revenue_each_year['year']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f' {year}', va='center', ha='left', color='black', fontsize=10)

plt.show()
```

```python
# 15 Horizontal Bar Charts with Year Labels on Top
# Highest Budget Movie Each Year

# Find the highest budget movie for each year
highest_budget_each_year = df.loc[df.groupby('year')['budget'].idxmax()][['year', 'movie_title', 'budget']]

# Plotting the highest budget movie each year
plt.figure(figsize=(12, 20))
bars = plt.barh(highest_budget_each_year['movie_title'], highest_budget_each_year['budget'], color='darkred')
plt.title('Highest Budget Movie Each Year')
plt.xlabel('Budget')
plt.ylabel('Movie Title')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest budget movie at the top

# Add year labels on top of the bars
for bar, year in zip(bars, highest_budget_each_year['year']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f' {year}', va='center', ha='left', color='black', fontsize=10)

plt.show()
```

```python
# 16
# Create Month vs Revenue Bar Plot

df['month'] = df['release_date'].dt.month

# Drop rows where release_date is NaN (if any)
df = df.dropna(subset=['release_date'])

# Aggregate data by month and calculate average revenue
month_revenue = df.groupby('month')['revenue'].mean()

# Plotting
plt.figure(figsize=(12, 6))
month_revenue.plot(kind='bar', color='grey')
plt.title('Average Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Average Revenue')

# Adjust xticks to show month names from 1 to 12
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
# 17 Visualize Cast in Most Movies - Use a bar plot to visualize the top actors and their frequencies.

# Split and explode the 'cast' column to count actor frequencies
cast_freq = df['cast'].str.split(',').explode().str.strip()
cast_counts = cast_freq.value_counts()

# Top 20 actors with the most appearances
top_cast = cast_counts.head(20)

# Plotting
plt.figure(figsize=(12, 8))
top_cast.sort_values().plot(kind='barh', color='purple')
plt.title('Top Actors by Number of Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Actor')
plt.gca().invert_yaxis()  # Invert y-axis to show the actor with the most movies on top
plt.show()
```
