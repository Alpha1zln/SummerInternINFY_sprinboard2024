```python
# ANS 3 - STEP 2 -
# CODE TO FETCH 2 COLS NULL VALUES FROM TMDB SITE USING API
# Fills missing movie_url and movie_title using the TMDB API.


import pandas as pd
import requests

# Function to fetch movie details from TMDB using IMDb ID
def fetch_movie_details(imdb_id, api_key):
    url = f"https://api.themoviedb.org/3/find/{imdb_id}"
    params = {
        'api_key': api_key,
        'external_source': 'imdb_id'
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data['movie_results']:
            movie = data['movie_results'][0]
            movie_url = f"https://www.themoviedb.org/movie/{movie['id']}"
            movie_title = movie['title']
            return movie_url, movie_title
    return None, None



# Your TMDB API key
api_key = '5abf4d126acc10d018d17b034dd7bb40'

# Iterate over the rows to fill missing movie_url and movie_title
for index, row in df.iterrows():
    if pd.isnull(row['movie_url']) or pd.isnull(row['movie_title']) or row['movie_url'] == '' or row['movie_title'] == '':
        imdb_id = row['imdb_id']
        movie_url, movie_title = fetch_movie_details(imdb_id, api_key)
        if movie_url:
            df.at[index, 'movie_url'] = movie_url
        if movie_title:
            df.at[index, 'movie_title'] = movie_title


```

```python
# ANS 3 - STEP 5 -
# "REST OF THE COLS" = NULL VALUES HANDLE

# Fills missing numerical values with the column's mean and categorical values with the column's mode.
# # Handling missing values for the rest of the columns


for column in df.columns:
    if df[column].isnull().sum() > 0 or (df[column] == '').sum() > 0:
        if df[column].dtype in ['float64', 'int64']:  # Numerical column
            mean_value = df[column].mean()
            df[column].fillna(mean_value, inplace=True)
            df[column].replace('', mean_value, inplace=True)
        else:  # Categorical column
            mode_value = df[column].mode()[0]
            df[column].fillna(mode_value, inplace=True)
            df[column].replace('', mode_value, inplace=True)


df.shape
```

4)check for any dataframes that has revenue as 0, average the revenue for the rest of dataframes and add it.


```python
# ANS 4 - STEP 1 -
# CHECK HOW MANY COLS HAVE REVENUE = 0

# Check how many rows have revenue as 0
zero_revenue_count = (df['revenue'] == 0).sum()
print(f"Number of rows with revenue = 0 are : {zero_revenue_count}")
```

```python
ANS 4 - STEP 2 -
REPLACE REV = 0 COLS , BY MEAN

# Calculate average revenue excluding rows where revenue is 0
average_revenue = df[df['revenue'] > 0]['revenue'].mean()

# didnt used int outside here... so will change float replaced to int later
# average_revenue = int(df[df['revenue'] > 0]['revenue'].mean())
# if had used, directly int above, then it would have replaced float vals to int

# Replace revenue 0 with the average revenue
df['revenue'] = df['revenue'].replace(0, average_revenue)

```

```python
# STEP 3 -
# CHECK IF REV=0 COLS ARE REPLACED BY MEAN


zero_revenue_count = (df['revenue'] == 0).sum()
print(f"Number of rows with revenue = 0 are : {zero_revenue_count}")
```
