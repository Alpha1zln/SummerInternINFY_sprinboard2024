# for COL =spoken_languages
# use one hot encoding for it

```python
# 2 COL = spoken_languages

unique_values_counts_spoken_languages = df['spoken_languages'].nunique()
print(f"Number of unique values in column 'spoken_languages': {unique_values_counts_spoken_languages}")

print('*')

# Find the unique values in the 'spoken_languages' column
unique_status_spoken_languages = df['spoken_languages'].unique()

# Print the unique values
print(f"Unique values in the 'spoken_languages' column: {unique_status_spoken_languages}")
```

```python
# Split the 'spoken_languages' column into a list of individual spoken_languages
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: x.split(', '))

# Initialize an empty set to store unique spoken_languages
unique_spoken_languages = set()

# Populate the set with unique spoken_languages from the 'spoken_languages' column
for spoken_languages_list in df['spoken_languages']:
    unique_spoken_languages.update(spoken_languages_list)

unique_spoken_languages
```

```python
# Create a new DataFrame to store binary columns for each unique spoken_language
spoken_languages_df = pd.DataFrame()

# Create binary columns for each unique spoken_language
for lang in unique_spoken_languages:
    spoken_languages_df[lang] = df['spoken_languages'].apply(lambda x: 1 if lang in x else 0)

# Concatenate the original DataFrame with the new binary columns DataFrame
df = pd.concat([df, spoken_languages_df], axis=1)
```
